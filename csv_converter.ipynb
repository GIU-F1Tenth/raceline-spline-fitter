# Converter from the map_converter script to pure x, y csv format

This is a Jupyter Notebook that converts the csv file from the map_converter script (header: 'x_m,y_m,w_tr_right_m,w_tr_left_m') into a cleaned csv for this package.


```python
import os
import csv
```

```python
CSV_FILE = "test_uncleaned.csv"  # The name of the csv file in the data folder
CLEANED_CSV_FILE = "cleaned.csv"  # The name of the cleaned csv file to be created
```

```python
# Sanity checks
if not os.path.exists(os.path.join("data", CSV_FILE)):
    raise FileNotFoundError(
        f"The file {CSV_FILE} does not exist. Please ensure it is in the current directory.")

if os.path.exists(os.path.join("data", CLEANED_CSV_FILE)):
    print(
        f"Warning: The file {CLEANED_CSV_FILE} already exists and will be overwritten.")
    input("Press Enter to continue or Ctrl+C to cancel...")
```

```python
csv_data = []

with open(os.path.join("data", CSV_FILE), 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        try:
            x = float(row[0])
            y = float(row[1])
            csv_data.append([x, y])
        except ValueError:
            continue  # Skip rows with non-numeric values
```

```python
with open(os.path.join("data", CLEANED_CSV_FILE), 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['x', 'y'])  # Write header
    writer.writerows(csv_data)
```
